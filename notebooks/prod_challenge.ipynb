```python
def get_results_in_file(file):
    step_1 = file['steps'][0]

    reactants = []
    for react in step_1['reactants']:
        reactants.append(react['smiles'])
    reagents = []
    for reagent in step_1['reagents']:
        reagents.append(reagent['smiles'])
    products = []
    for prod in step_1['products']:
        products.append(prod['smiles'])
    return reactants, reagents, products
```

```python
# actual evaluation
from rdkit import Chem


def actual_eval(df_results):
    # check if the predicted outputs are in the actual outputs
    # and if the number of predicted outputs is equal to the number of actual outputs
    not_common = []
    missed = []
    for i, row in df_results.iterrows():
        outputs = row['output'].split('.')
        # print(outputs)
        predicted_outputs = row['reactants'] + row['reagents']
        # print(outputs, predicted_outputs)
        outputs_mols = [Chem.CanonSmiles(output) for output in outputs]
        predicted_outputs_mols = [
            Chem.CanonSmiles(output) for output in predicted_outputs
        ]
        # print(outputs_mols,predicted_outputs_mols)
        # break
        # check if all the predicted outputs are in the actual outputs
        if any([output in outputs_mols for output in predicted_outputs_mols
                ]):
            df_results.loc[i, 'any_correct'] = 1
        else:
            df_results.loc[i, 'any_correct'] = 0

        if all([output in outputs_mols for output in predicted_outputs_mols
                ]):
            df_results.loc[i, 'all_correct'] = 1
        else:
            df_results.loc[i, 'all_correct'] = 0
        # get the molecules that are not common
        not_common.append([
            output for output in predicted_outputs_mols
            if output not in outputs_mols
        ])
        # df_results.loc[i, 'not_common'] =
        # get the molecules that are missed
        missed.append([
            output for output in outputs_mols
            if output not in predicted_outputs_mols
        ])

    df_results['not_common'] = not_common
    df_results['missed'] = missed
    all_correct =  100 * df_results.all_correct.sum() / len(df_results)
    any_correct =  100 * df_results.any_correct.sum() / len(df_results)
    return all_correct, any_correct, df_results
```

```python
import ast
def resolve_str(mol):
    if type(mol) == str:
        return ast.literal_eval(mol)
    else:
        return mol
```

```python
# load up the results
import json
import os
import pandas as pd
import ast

dir_results_main = "../results/benchmarking/AZ/"


# folder_list = ["m0p1:Pistachio_100+", "m1p0:Pistachio_100+","m0p1:USPTO", "m1p0:USPTO" ]
folder_list = ["Pistachio_100+", "USPTO"]

for fol in folder_list:
    dir_results = f"{dir_results_main}/{fol}/"
    # find folders in dir_results
    for run in range(0,1):
        # try:
        if True:
            dir_results = f"{dir_results}/run_{run}/"

            # find folders in dir_results
            files = [
                f for f in os.listdir(dir_results)
                if os.path.isfile(os.path.join(dir_results, f))
            ]

            # load up the results in a df
            results = {}
            for file in files:
                with open(os.path.join(dir_results, file)) as f:
                    try:
                        reactants, reagents, products = get_results_in_file(json.load(f))
                        # results[df_dict[file.split(".js")[0]]] = {
                        results[int(file.split(".js")[0])] = {
                            "reactants": reactants,
                            "reagents": reagents,
                            "products": products
                        }
                    except Exception as e:
                        print("Error in file:", file)
                        print("Error:", e)
                        continue
            results_df = pd.DataFrame(results).T
            # sort based on index
            results_df.sort_index(inplace=True)
            # results_df.to_csv(f"../results/USPTO_100/{fol}/results_run_{run}.csv")
            # results_df = pd.read_csv(f"../results/USPTO_100/{fol}/results_run_{run}_full.csv", index_col=0)
            # print(results_df.describe())
            df_results = df.join(results_df, on=df.index)
            
            
            df_results.dropna(inplace=True)

            # convert the smiles to a list
            df_results['reactants'] = df_results['reactants'].apply(
                lambda x: resolve_str(x))
            df_results['reagents'] = df_results['reagents'].apply(
                lambda x: resolve_str(x))
            df_results['products'] = df_results['products'].apply(
                lambda x: resolve_str(x))
            df_results.to_csv(f"{dir_results}/results_run_{run}_full2.csv")
            all_correct, any_correct, df_res_final = actual_eval(df_results)
            df_res_final.to_csv(f"{dir_results}/results_run_{run}_final.csv")
            print(f"All correct {fol} - run {run}: {all_correct}")
            print(f"Any correct {fol} - run {run}: {any_correct}")
        # except Exception as e:
        else:
            print("Error in folder:", fol)
            print("Error:", e)
        
```

```python
import ast
import pandas as pd

df_results = pd.read_csv("data/USPTO-50k_500_results.csv", index_col=0)
# drop nan values
df_results.dropna(inplace=True)

# convert the smiles to a list
df_results['reactants'] = df_results['reactants'].apply(
    lambda x: ast.literal_eval(x))
df_results['reagents'] = df_results['reagents'].apply(
    lambda x: ast.literal_eval(x))
df_results['products'] = df_results['products'].apply(
    lambda x: ast.literal_eval(x))
df_results
```

```python
# actual evaluation
from rdkit import Chem

not_common = []
missed = []

for i, row in df_results.iterrows():
    outputs = row['output'].split('.')
    predicted_outputs = row['reactants'] + row['reagents']
    outputs_mols = [Chem.CanonSmiles(output) for output in outputs]
    predicted_outputs_mols = [
        Chem.CanonSmiles(output) for output in predicted_outputs
    ]
    # print(outputs_mols)
    # print(predicted_outputs_mols)
    # break
    # check if all the predicted outputs are in the actual outputs
    if any([output in outputs_mols for output in predicted_outputs_mols
            ]) and len(outputs_mols) == len(predicted_outputs_mols):
        df_results.loc[i, 'any_correct'] = 1
    else:
        df_results.loc[i, 'any_correct'] = 0

    if all([output in outputs_mols for output in predicted_outputs_mols
            ]) and len(outputs_mols) == len(predicted_outputs_mols):
        df_results.loc[i, 'all_correct'] = 1
    else:
        df_results.loc[i, 'all_correct'] = 0
    # get the molecules that are not common
    not_common.append([
        output for output in predicted_outputs_mols
        if output not in outputs_mols
    ])
    # df_results.loc[i, 'not_common'] =
    # get the molecules that are missed
    missed.append([
        output for output in outputs_mols
        if output not in predicted_outputs_mols
    ])

df_results['not_common'] = not_common
df_results['missed'] = missed
```
